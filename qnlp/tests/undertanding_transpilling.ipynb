```python
import qiskit as qk
from context import tools as t
```

```python
qr = qk.QuantumRegister(2, 'q')
qc = qk.QuantumCircuit(qr)
qc.swap(0, 1)
```

```python
qr = qk.QuantumRegister(3, 'q')
qc = qk.QuantumCircuit(qr)
qc.cswap(0, 1, 2)
```

```python
print(qc.draw())
```

```python
print(qc.draw(output='mpl'))
```
