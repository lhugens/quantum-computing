```python
import qiskit as qk
```

```python
qk.IBMQ.load_account()
```

```python
provider = qk.IBMQ.get_provider(hub='ibm-q')
ibm = lambda x:(
    x.status().operational and
    not x.configuration().simulator and
    x.configuration().n_qubits > 1
)
sim = lambda x:(
    x.status().operational and
    x.configuration().simulator and
    x.configuration().n_qubits > 1
)
backends_ibm = provider.backends(filters=ibm)
backends_sim = provider.backends(filters=sim)
```

```python
bogota = backends_ibm[0]
```

```python
bogota.properties().to_dict()
```

```python
l = ['name', 'max_jobs', 'max_experiments', 'max_shots', 'num_qubits']
print(f'{l[0]: <15} {l[1]: >8} {l[2]: >2} {l[3]: >3} {l[4]: >3}')

for b in backends_ibm:
    c = b.configuration()
    print(f'{c.backend_name: <15} {b.job_limit().maximum_jobs: >8} {c.max_experiments: >15} {c.max_shots: >9} {c.num_qubits: >10}')
```
